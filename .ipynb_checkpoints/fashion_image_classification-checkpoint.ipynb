# Model Building

**Step 1: Normalize the Pixel Values**

```python
# Normalize the images
x_train = x_train / 255.0
x_test = x_test / 255.0
```

**Step 2: Reshape for CNN**

```python
# Add a channel dimension: (60000, 28, 28) → (60000, 28, 28, 1)
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)
```

**Step 3: One-hot encode the labels**

```python
from tensorflow.keras.utils import to_categorical

y_train_cat = to_categorical(y_train, num_classes=10)
y_test_cat = to_categorical(y_test, num_classes=10)
```

**Step 4: Train-Validation Split**

```python
from sklearn.model_selection import train_test_split

x_train_final, x_val, y_train_final, y_val = train_test_split(
    x_train, y_train_cat, test_size=0.1, random_state=42
)
```

**Step 5: Build a CNN model**

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    MaxPooling2D(2, 2),
    
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),
    
    Flatten(),
    Dropout(0.25),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])
```

**Step 6: Compile the model**

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

**Step 7: Train the model**

```python
history = model.fit(
    x_train_final, y_train_final,
    validation_data=(x_val, y_val),
    epochs=10,
    batch_size=64
)
```

**Train vs Validation Accuracy**

```python
plt.plot(history.history['accuracy'], label='train_acc')
plt.plot(history.history['val_accuracy'], label='val_acc')
plt.legend()
```

This plot shows how the model's accuracy evolved over each epoch for both training and validation sets. It helps detect overfitting or underfitting. 

## Prediction on Unseen Data

```python
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)
print(f"\nTest accuracy: {test_acc:.4f}")
```

**Correctly classified Images - Sample**

```python
# Predict
predictions = model.predict(x_test)
predicted_labels = np.argmax(predictions, axis=1)

# Plot 10 random test images with predicted & true labels
indices = np.random.choice(len(x_test), 10, replace=False)

plt.figure(figsize=(8, 4))
for idx, i in enumerate(indices):
    plt.subplot(2, 5, idx + 1)
    plt.imshow(x_test[i], cmap='gray')
    plt.title(f"P: {label_names[predicted_labels[i]]}\nA: {label_names[y_test[i]]}", fontsize=8)
    plt.axis('off')

plt.tight_layout()
plt.show()
```

**Misclassified Images - Sample**

```python
y_true = y_test.flatten()

# Find misclassified indices
misclassified_idx = np.where(predicted_labels != y_true)[0]

# Track first misclassified sample for each true class
seen_labels = set()
misclassified_samples = []

for idx in misclassified_idx:
    label = y_true[idx]
    if label not in seen_labels:
        misclassified_samples.append(idx)
        seen_labels.add(label)
    if len(seen_labels) == 10:
        break


plt.figure(figsize=(8, 4))
for i, idx in enumerate(misclassified_samples):
    plt.subplot(2, 5, i + 1)
    plt.imshow(x_test[idx], cmap='gray')
    plt.title(f"True: {label_names[y_true[idx]]}\nPred: {label_names[predicted_labels[idx]]}")
    plt.axis('off')
plt.tight_layout()
plt.show()
```
